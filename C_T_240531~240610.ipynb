```python
class Solution {
    public int solution(String before, String after) {
        int beforeSum = 0;
        int afterSum = 0;
        int rslt = 0;

        if(after.length() != before.length()) return rslt;

        for(int i = 0; i < after.length(); i++) {
            beforeSum += before.charAt(i);
            afterSum += after.charAt(i);
        }

        if(beforeSum == afterSum) rslt=1;

        return rslt;
    }
}
```

네 가지를 붙여 만들 수 있는 발음 이외에는 어떤 발음도 할 수 없는 것으로 규정합니다. 예를 들어 "woowo"는 "woo"는 발음할 수 있지만 "wo"를 발음할 수 없기 때문에 할 수 없는 발음입니다.

```python
class Solution {
    public int solution(String[] babbling) {
        int answer = 0;
        String[] ongal = {"aya","ye","woo","ma"};
        String temptxt = null;

        for(int i=0; i<babbling.length; i++){
            temptxt = babbling[i];
            for(int j=0; j<ongal.length; j++){
                if(temptxt.equals(babbling[i])){
                    temptxt = babbling[i].replaceAll(ongal[j],"0");
                }else{
                    temptxt = temptxt.replaceAll(ongal[j],"0");
                }
            }

            if(temptxt.replaceAll("0","").equals("")) answer++;
        }
        return answer;
    }
}
```

0 ≤ k ≤ 9
입출력 예

3부터 10까지 2는 한 번도 등장하지 않으므로 0을 return 합니다.

```python
class Solution {
    public int solution(int i, int j, int k) {
        int answer = checkNum(i,j,k);
        return answer;
    }

    private int checkNum(int i, int j, int k){
        int count = 0;

        while(true){

            String tmp1 = String.valueOf(i);
            String kStr = String.valueOf(k);
            String[] tmpSplit = tmp1.split("");
            for(int h = 0; h < tmpSplit.length; h++){
                if(tmpSplit[h].contains(kStr)){
                    count++;
                }
            }

            if(i == j){
                break;
            }

            i++;
        }

        return count;
    }
}
```

```python
class Solution {
    public int solution(int[] common) {
        int answer = 0;
        String memo = "";
        int a = common[1] - common[0];
        int b = common[2] - common[1];
        int len = common.length-1;

        boolean isGeometric = false;

        if(a != b && ( b%a == 0)){
            if(a < 0 || b < 0){
                memo = "minusGeometric";
            }
            isGeometric = true;
        }

        if(isGeometric){
            if(common[0] != 1){
                answer = common[len] * a;
            }else {
                answer = common[len] * b;
            }
        }else {
            if(common[0] == 0){
                answer = common[len] + b;
            }else {
                answer = common[len] + a;
            }
        }

        if(!memo.equals("")){
            answer = common[len] * (common[1]*common[0]);
        }
        return answer;

        // common[1]-common[0] == common[2] - common[1] 등차
        // common[1]-common[0] =! common[2] - common[1] && common[2]-common[1]%common[1]-common[0] = 0 등비수열
    }
}
```

```python
class Solution {
    public int[] solution(int num, int total) {
        boolean isOdd = true;
        int centerNum = 0;
        if(num%2 == 0){
            isOdd = false;
        }

        if(isOdd){
            centerNum = total/num;
        }else {
            centerNum = total/num +1;
        }

        int[] answer = new int[num];

        for(int i = 0; i < num; i++){
            answer[i] = (centerNum-(num/2)+i);
        }

        return answer;
    }
}
```
