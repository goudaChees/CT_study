```python
class Solution {
    public int solution(String before, String after) {
        int beforeSum = 0;
        int afterSum = 0;
        int rslt = 0;

        if(after.length() != before.length()) return rslt;

        for(int i = 0; i < after.length(); i++) {
            beforeSum += before.charAt(i);
            afterSum += after.charAt(i);
        }

        if(beforeSum == afterSum) rslt=1;

        return rslt;
    }
}
```

유의사항
네 가지를 붙여 만들 수 있는 발음 이외에는 어떤 발음도 할 수 없는 것으로 규정합니다. 예를 들어 "woowo"는 "woo"는 발음할 수 있지만 "wo"를 발음할 수 없기 때문에 할 수 없는 발음입니다.

```python
class Solution {
    public int solution(String[] babbling) {
        int answer = 0;
        String[] ongal = {"aya","ye","woo","ma"};
        String temptxt = null;

        for(int i=0; i<babbling.length; i++){
            temptxt = babbling[i];
            for(int j=0; j<ongal.length; j++){
                if(temptxt.equals(babbling[i])){
                    temptxt = babbling[i].replaceAll(ongal[j],"0");
                }else{
                    temptxt = temptxt.replaceAll(ongal[j],"0");
                }
            }

            if(temptxt.replaceAll("0","").equals("")) answer++;
        }
        return answer;
    }
}
```

머쓱이는 프로그래머스에 로그인하려고 합니다. 머쓱이가 입력한 아이디와 패스워드가 담긴 배열 id_pw와 회원들의 정보가 담긴 2차원 배열 db가 주어질 때, 다음과 같이 로그인 성공, 실패에 따른 메시지를 return하도록 solution 함수를 완성해주세요.

아이디와 비밀번호가 모두 일치하는 회원정보가 있으면 "login"을 return합니다.
로그인이 실패했을 때 아이디가 일치하는 회원이 없다면 “fail”를, 아이디는 일치하지만 비밀번호가 일치하는 회원이 없다면 “wrong pw”를 return 합니다.

```python
import java.util.*;
import java.lang.*;

class Solution {
    public String solution(String[] id_pw, String[][] db) {
        String answer = "";
        String id = id_pw[0];
        String pw = id_pw[1];

        for(int i = 0; i < db.length; i++){
            for(int j =0; j < 1; j++){
                if(id.equals(db[i][j])){
                    answer = "idPass";
                    if(pw.equals(db[i][j+1])){
                        answer = "login";
                    }else{
                        answer = "wrong pw";
                    }
                }else {
                    answer = "notMatch";
                }
            }

        }

        if(answer.equals("notMatch")){
            answer = "fail";
        }


        return answer;
    }
}
```
