```python
## 일부 정답

import java.util.*;
import java.lang.*;

class Solution {
    public String solution(String polynomial) {
        String answer = "";
        String[] split = polynomial.split(" \\+ ");

        List<String> listForX = new ArrayList<>();
        List<Integer> listForNum = new ArrayList<>();

        int a = 0;
        int b = 0;
        String c = "";


        // x식과 상수식 분리
        for (int i = 0; i < split.length; i++){
            if(split[i].contains("x")){
                listForX.add(split[i]);
            }else{
                int nums = Integer.parseInt(split[i]);
                listForNum.add(nums);
            }
        }

        // x식의 값 계산
        for (int i = 0; i < listForX.size(); i++){
            String tmp = listForX.get(i).replace("x","");
            c += tmp;
        }

        if (c.length() == 0){
            a = listForX.size();
        }else {
            int tmp = listForX.size() - c.length();
            if(c.length() > 1){
                for(int i = 0; i < c.length(); i++){
                    char ch = c.charAt(i);
                    tmp += Character.getNumericValue(ch);
                }
                a = tmp;
            }else {
                tmp += Integer.parseInt(c);
                a = tmp;
            }
        }



        // 상수 값 계산
        if(listForNum.size() != 0 ){
            for (int i = 0; i < listForNum.size(); i++){
                b += listForNum.get(i);
            }
            answer = String.valueOf(a)+"x + ";
            answer += String.valueOf(b);
        }else {
            answer = String.valueOf(a)+"x";
        }


        return answer;
    }
}
```

```python
## 자바 정답 1

class Solution {
        public int solution(int[][] dots) {
            int answer = 0;

            // 1,2 번째 점
            int a = (dots[0][1] - dots[1][1]) * (dots[2][0] - dots[3][0]);
            int b = (dots[2][1] - dots[3][1]) * (dots[0][0] - dots[1][0]);
            if(a == b || (Math.abs(a) == 1 && Math.abs(b)==1)) return 1;

            //1,3 번째 점
            a = (dots[0][1] - dots[2][1]) * (dots[1][0] - dots[3][0]);
            b = (dots[1][1] - dots[3][1]) * (dots[0][0] - dots[2][0]);
            if(a == b || (Math.abs(a) == 1 && Math.abs(b)==1)) return 1;

            // 1,4 번째 점
            a = (dots[0][1] - dots[3][1]) * (dots[2][0] - dots[1][0]);
            b = (dots[2][1] - dots[1][1]) * (dots[0][0] - dots[3][0]);
            if(a == b || (Math.abs(a) == 1 && Math.abs(b)==1)) return 1;

            return answer;
        }
    }
```

```python
## 자바 정답 2222

import java.lang.Math;
class Solution {
    private int xPos = 0;
    private int yPos = 1;
    private int[][] checkCases = {{0, 1, 2, 3}, {0, 2, 1, 3}, {0, 3, 1, 2}};
    public int solution(int[][] dots) {
        int answer = 0;
        for (int[] checkCase : checkCases) {
            if (checkParallel(checkCase, dots)) {
                answer = 1;
                break;
            }
        }

        return answer;
    }

    private boolean checkParallel(int[] checkCase, int[][] dots) {
        int xPos1 = dots[checkCase[0]][xPos];
        int yPos1 = dots[checkCase[0]][yPos];
        int xPos2 = dots[checkCase[1]][xPos];
        int yPos2 = dots[checkCase[1]][yPos];
        int xPos3 = dots[checkCase[2]][xPos];
        int yPos3 = dots[checkCase[2]][yPos];
        int xPos4 = dots[checkCase[3]][xPos];
        int yPos4 = dots[checkCase[3]][yPos];
        // (y4 - y3)(x2 - x1) - (x4 - x3)(y2 - y1) 평행여부 조건
        int chk = ((yPos4 - yPos3) * (xPos2 - xPos1)) - ((xPos4 - xPos3) * (yPos2 - yPos1));
        if (chk == 0) {
            return true;
        }
        return false;
    }
}
```

```python
## 자바 정답 1

import java.util.HashMap;
import java.util.Map;
class Solution {
    public int solution(int[][] lines) {
        Map<Integer, Integer> map = new HashMap<>();

        for (int i=0; i<lines.length; i++) {
            int min = Math.min(lines[i][0], lines[i][1]);
            int max = Math.max(lines[i][0], lines[i][1]);

            for (int j=min; j<max; j++) {
                map.put(j, map.getOrDefault(j, 0) + 1);
            }
        }

        int answer = 0;

        for (Map.Entry<Integer, Integer> entry : map.entrySet()) {
            if (entry.getValue() >= 2) {
                answer++;
            }
        }

        return answer;
    }
}
```

```python
## 자바 정답 4

class Solution {
    public int solution(int a, int b) {
        int gcd = getGcd(a, b);
        b /= gcd;

        while (b % 5 == 0) {
            b /= 5;
        }

        while (b % 2 == 0) {
            b /= 2;
        }

        return b == 1 ? 1 : 2;
    }

    private int getGcd(int a, int b) {
        int temp;
        if (a < b) { // a < b의 경우 swap
            temp = b;
            b = a;
            a = temp;
        } else if (a == b) {
            return a;
        }

        while (b != 0) { // a > b 가정
            temp = a % b;
            a = b;
            b = temp;
        }

        return a;
    }
}
```

```python
## 자바 정답 2

import java.util.*;
import java.lang.Math;
class Solution {
    public int[] solution(int[] numlist, int n) {
        int[] answer = new int[numlist.length];
        int idx = 0;
        Map<Integer, List<Integer>> numMap = new HashMap<Integer, List<Integer>>();
        for (int num : numlist) {
            int key = Math.abs(n - num);
            List<Integer> list;
            if (numMap.containsKey(key)) {
                list = numMap.get(key);
            } else {
                list = new ArrayList<Integer>();
            }
            list.add(num);
            numMap.put(key, list);
        }
        List<Integer> keyList = new ArrayList(numMap.keySet());
        Collections.sort(keyList);
        for (int key : keyList) {
            List<Integer> list = numMap.get(key);
            Collections.sort(list, Collections.reverseOrder());
            for (int num : list) {
                answer[idx++] = num;
            }
        }
        return answer;
    }
}
```

```python
## 자바 정답 3

import java.util.Arrays;

class Solution {
    public int[] solution(int[] numList, int n) {
        return Arrays.stream(numList)
                .boxed()
                .sorted((a, b) -> Math.abs(a - n) == Math.abs(b - n) ? b.compareTo(a) : Integer.compare(Math.abs(a - n), Math.abs(b - n)))
                .mapToInt(Integer::intValue)
                .toArray();
    }
}
```
