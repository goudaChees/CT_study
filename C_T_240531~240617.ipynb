```python
class Solution {
    public int solution(String before, String after) {
        int beforeSum = 0;
        int afterSum = 0;
        int rslt = 0;

        if(after.length() != before.length()) return rslt;

        for(int i = 0; i < after.length(); i++) {
            beforeSum += before.charAt(i);
            afterSum += after.charAt(i);
        }

        if(beforeSum == afterSum) rslt=1;

        return rslt;
    }
}
```

```python
class Solution {
    public int solution(String[] babbling) {
        int answer = 0;
        String[] ongal = {"aya","ye","woo","ma"};
        String temptxt = null;

        for(int i=0; i<babbling.length; i++){
            temptxt = babbling[i];
            for(int j=0; j<ongal.length; j++){
                if(temptxt.equals(babbling[i])){
                    temptxt = babbling[i].replaceAll(ongal[j],"0");
                }else{
                    temptxt = temptxt.replaceAll(ongal[j],"0");
                }
            }

            if(temptxt.replaceAll("0","").equals("")) answer++;
        }
        return answer;
    }
}
```

```python
class Solution {
    public int solution(int i, int j, int k) {
        int answer = checkNum(i,j,k);
        return answer;
    }

    private int checkNum(int i, int j, int k){
        int count = 0;

        while(true){

            String tmp1 = String.valueOf(i);
            String kStr = String.valueOf(k);
            String[] tmpSplit = tmp1.split("");
            for(int h = 0; h < tmpSplit.length; h++){
                if(tmpSplit[h].contains(kStr)){
                    count++;
                }
            }

            if(i == j){
                break;
            }

            i++;
        }

        return count;
    }
}
```

```python
class Solution {
    public String solution(String my_string, int[][] queries) {
        String answer = my_string;
        for(int i = 0; i < queries.length; i++){
            answer = reverseStr(answer, queries[i][0], queries[i][1]);
        }
        return answer;
    }

    protected String reverseStr(String my_string, int startNum, int endNum){
        String[] split = my_string.split("");
        String answer = "";
        for(int i = 0; i < startNum; i++){
            answer += split[i];
        }

        for(int j = endNum; j >= startNum; j--){
            answer += split[j];
        }

        for(int k = endNum+1; k < split.length; k++){
            answer += split[k];
        }
        return answer;
    }
}
```

```python
import java.util.*;

class Solution {
    public String[] solution(String myString) {
        ArrayList list = new ArrayList();
        String replaceStr = "";
        String[] split = myString.split("");

        for(int i = 0; i < split.length; i++){
            if(split[i].equals("x")){
                if(!replaceStr.equals("")){
                    list.add(replaceStr);
                }
                replaceStr = "";
            }else {
                replaceStr += split[i];
            }

            if(i == split.length-1){
                if(!split[i].equals("x")){
                    list.add(replaceStr);
                }
            }
        }

        String[] answer = new String[list.size()];

        for(int j = 0; j < list.size(); j++){
            answer[j] = (String) list.get(j);
        }

        Arrays.sort(answer);

        return answer;
    }
}
```

```python
import java.util.*;
import java.lang.*;

class Solution {
    public int[] solution(int[] arr, int[] query) {
        ArrayList copyArr = new ArrayList();
        for(int m : arr){
            copyArr.add(m);
        }

        for(int i = 0; i< query.length; i++){
            int size = copyArr.size();
            if(i%2 == 0){
                for(int j = size-1; j > query[i]; j--){
                    copyArr.remove(j);
                }
            }else {
                for(int k = 0; k < query[i]; k++){
                    copyArr.remove(k);
                }
            }
        }

        int[] answer = new int[copyArr.size()];

        for(int l = 0; l < copyArr.size(); l++){
            answer[l] = (int)copyArr.get(l);
        }

        return answer;

    }
}
```

```python
import java.util.*;

class Solution {
    public int solution(String[] strArr) {
        HashMap<Integer, Integer> map = new HashMap<Integer, Integer>();

        for(String s : strArr){
            if(map.get(s.length()) == null){
                map.put(s.length(), 1);
            }else {
                map.put(s.length(), map.get(s.length())+1);
            }
        }

        int answer = 0;

        for(Integer key :  map.keySet()){

            if(map.get(key) > answer){
                answer = map.get(key);
            }
        }

        return answer;
    }
}
```

```python
## Stack으로 풀어내는 방법
##

import java.util.Stack;

class Solution {
    public int[] solution(int[] arr) {

        Stack<Integer> stack = new Stack<>();

        for (int no : arr) {
            if (!stack.isEmpty() && no == stack.peek()) {
                stack.pop();
            } else {
                stack.push(no);
            }
        }

        return stack.isEmpty() ? new int[] { -1 } : stack.stream().mapToInt(i -> i).toArray();
    }
}
```
